```python
import pandas as pd

car_data_raw = pd.read_csv("https://cdn.c18l.org/vehicles_lab.csv")
```

```python
import numpy as np
```

```python
car_data_raw.head()
```

```python
car_data_raw.describe()
```

```python
small_data = car_data_raw.loc[:, [
    'price', 'condition', 'odometer',
    'year', 'size', 'type', 'title_status'
]]

small_data.head()
```

```python
temp = pd.get_dummies(small_data['condition'], dummy_na=True).astype(int)

temp.head()
```

```python
new_data = pd.concat([small_data, temp], axis=1)
new_data.head()
```

```python
temp = pd.get_dummies(small_data['size'], dummy_na=True).astype(int)
new_data = pd.concat([new_data, temp], axis=1)
temp.head()



```

```python
temp = pd.get_dummies(small_data['type'], dummy_na=True).astype(int)
new_data = pd.concat([new_data, temp], axis=1)
temp.head()
```

```python
temp = pd.get_dummies(small_data['title_status'], dummy_na=True).astype(int)
new_data = pd.concat([new_data, temp], axis=1)
temp.head()
```

```python
new_data.head()
```

```python
pd.set_option('display.max_columns', None)
new_data.head()
```

```python
new_clean_data = new_data.drop(columns = ['title_status', 'type', 'condition', 'size'],axis=1)
new_clean_data.head()
```

```python
new_clean_data.describe()
```

```python
new_clean_data = new_clean_data.loc[small_data['price'] <= 30000.0]
new_clean_data = new_clean_data.loc[small_data['odometer'] <= 50000.0]
new_clean_data = new_clean_data.loc[small_data['price'] > 2500]
new_clean_data = new_clean_data.loc[small_data['odometer'] > 12000]
new_clean_data = new_clean_data.loc[small_data['year'] >= 1970.0]
new_clean_data.describe()


```

```python
# car_data_raw['posting_year'] = car_data_raw['posting_date'].str.extract(r'^([0-9]{4})').astype(int)

new_clean_data['car_age'] = 2023 - new_clean_data['year']
new_clean_data = new_clean_data[new_clean_data['car_age'] >= 0]



new_clean_data['miles_per_year'] = new_clean_data['odometer'] / new_clean_data['car_age']
new_clean_data = new_clean_data[new_clean_data['car_age'] >= 0]
new_clean_data.describe()

```

```python
new_clean_data['binned_price']= pd.cut(
    x = new_clean_data['price'],
    bins = np.linspace(0, 50000, 5+1),
    labels = False
)
new_clean_data = new_clean_data.dropna()
new_clean_data.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(
    new_clean_data,
    train_size=0.8,
    random_state=42)

model = LogisticRegression(multi_class='multinomial', max_iter = 10000).fit(
    X=train_data.loc[:, [
        'odometer',	'year',
        'excellent',	'fair',	'good',
        'like new',	'new',	'salvage',
        'compact',	'full-size',	'mid-size',
        'sub-compact',	'SUV',	'bus',
        'convertible',	'coupe',	'hatchback',
        'mini-van',	'offroad',	'other',
        'pickup',	'sedan',	'truck',
        'van',	'wagon', 'clean',	'lien',
        'missing',	'rebuilt',
        'salvage',	'car_age',	'miles_per_year']],
    y=train_data['binned_price']
)
```
